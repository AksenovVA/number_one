* __Задача 5__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ – (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import time
import functions
import functions_vectorized

from importlib import reload
reload(functions)
reload(functions_vectorized)

cnt = 100

Ox = list()
Oy = list()
Oxv = list()
Oyv = list()

for k in range(10, 1000):
    sm1 = 0
    sm2 = 0
    mas = np.random.randint(0, 5, size = k)
    for i in range(cnt):
        
        start1 = time.time()
        k1 = functions.run_length_encoding(mas)
        end1 = time.time()
        start2 = time.time()
        k2 = functions_vectorized.run_length_encoding(mas)
        
        end2 = time.time()
        sm1 += end1 - start1
        sm2 += end2 - start2
    Ox.append(k)
    Oy.append(sm1 / cnt)
    Oxv.append(k)
    Oyv.append(sm2 / cnt)


plt.plot(Ox, Oy, label=u"Невекторизированная")
plt.plot(Oxv, Oyv, label=u"Векторизированная")
plt.title("График")

plt.xlabel("Длина массива")
plt.ylabel("Время (с)")
plt.grid()
plt.legend()
plt.show()
```

 
 
* __Задача 6__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Дополнительно сравните с функцией scipy.spatial.distance.cdist по скорости работы (сравнения приведите ниже в ноутбуке).

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
import random
import time
import functions
import functions_vectorized

from importlib import reload
reload(functions)
reload(functions_vectorized)

cnt = 50

Ox = list()
Oy = list()
Oxv = list()
Oyv = list()
Oxx = list()
Oyy = list()

for i in range(10, 100):
    for j in range(i, 100):
        x = np.random.rand(i, j)
        y = np.random.rand(i, j)
        sm1 = 0
        sm2 = 0
        sm3 = 0
        for m in range(cnt):
            start1 = time.time()
            k1 = functions.pairwise_distance(x, y)

            end1 = time.time()
            start2 = time.time()
            k2 = functions_vectorized.pairwise_distance(x, y)
            end2 = time.time()
            start3 = time.time()
            k3 = scipy.spatial.distance.cdist(x, y)
            end3 = time.time()
            sm1 += end1 - start1
            sm2 += end2 - start2
            sm3 += end3 - start3
            
            
        Ox.append(i)
        Oy.append(sm1 / cnt)
        Oxv.append(i)
        Oyv.append(sm2 / cnt)
        Oxx.append(i)
        Oyy.append(sm3 / cnt)


plt.plot(Ox, Oy, label=u"Невекторизированная")
plt.plot(Oxv, Oyv, label=u"Векторизированная")
plt.plot(Oxx, Oyy, label=u"Встроенная")
plt.title("График")

plt.xlabel("Длина диагонали")
plt.ylabel("Время (с)")
plt.grid()
plt.legend()
plt.show()
```
